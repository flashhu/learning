### 2.4 数据的合并

#### 2.4.1 任务一：将data文件夹里面的所有数据都载入，观察数据的之间的关系

```python
# 载入data文件中的:train-left-down.csv
df_1eft_down = pd.read_csv('./data/train-left-down.csv')
df_1eft_down.head()
```

```python
# 载入data文件中的:train-right-up.csv
df_right_up = pd.read_csv('./data/train-right-up.csv')
df_right_up.head()
```

```python
# 载入data文件中的:train-right-down.csv
df_right_down = pd.read_csv('./data/train-right-down.csv')
df_right_down.head()
```

【提示】结合之前我们加载的train.csv数据，大致预测一下上面的数据是什么

* left, right: 原数据集的左/右边部分
* up，down：原数据集的前/后半部分

#### 2.4.2：任务二：使用concat方法：将数据train-left-up.csv和train-right-up.csv横向合并为一张表，并保存这张表为result_up

```python
result_up = pd.concat([df_1eft_up, df_right_up], axis=1)
result_up.head()
```

#### 2.4.3 任务三：使用concat方法：将train-left-down和train-right-down横向合并为一张表，并保存这张表为result_down。然后将上边的result_up和result_down纵向合并为result。

```python
result_down = pd.concat([df_1eft_down, df_right_down], axis=1)
result_down.head()
```

```python
result = pd.concat([result_up, result_down], ignore_index=True)
result
```

#### 2.4.4 任务四：使用DataFrame自带的方法join方法和append：完成任务二和任务三的任务

```python
# join  左右合
result_up_j = df_1eft_up.join(df_right_up)
result_up_j.head()
```

```python
result_down_j = df_1eft_down.join(df_right_down)
result_down_j.head()
```

```python
# append 上下合
result_a = result_up_j.append(result_down_j, ignore_index=True)
result_a
```

#### 2.4.5 任务五：使用Panads的merge方法和DataFrame的append方法：完成任务二和任务三的任务

```python
# merge 左右合
result_up_m = pd.merge(df_1eft_up, df_right_up, left_index=True, right_index=True)
result_up_m.head()
```

```python
result_down_m = pd.merge(df_1eft_down, df_right_down, left_index=True, right_index=True)
result_down_m.head()
```

```python
# append 上下合
result_a = result_up_m.append(result_down_m, ignore_index=True)
result_a
```

【思考】对比merge、join以及concat的方法的不同以及相同。思考一下在任务四和任务五的情况下，为什么都要求使用DataFrame的append方法，如何只要求使用merge或者join可不可以完成任务四和任务五呢？

#### 2.4.6 任务六：完成的数据保存为result.csv
